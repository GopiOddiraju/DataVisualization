# **Q1**

```python
import pandas as pd

df = pd.read_csv('10s0118.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sorted_df = df.sort_values(by='Total', ascending=False)
plt.figure(figsize=(10, 10))
ax=sns.barplot(x='Total', y='State', data=sorted_df, color='purple')

plt.xlabel('Death rates per 100,000 population')
plt.ylabel('State')
plt.yticks(fontsize=8)
plt.xticks([0,100,200,300,400,500,600,700,800,900,1000])
plt.title('State-wise Death Rates per 100,000 population in the United States in 2006')
plt.show()
```

# **Q2**

```python

fig, ax = plt.subplots(figsize=(12, 20))
sorted_by_accidents = df.sort_values(by='Deaths due to Accidents,Injury by firearms,Assault', ascending=False)
bars_diseases = ax.barh(sorted_by_accidents['State'], sorted_by_accidents['Deaths due to Diseases'], color='purple',height=0.5, label='Death rates due to Diseases')
bars_accidents = ax.barh(sorted_by_accidents['State'], -sorted_by_accidents['Deaths due to Accidents,Injury by firearms,Assault'], color='green',height=0.5, label='Death rates due to Accidents, Injury by firearms, Assault')

ax.set_xlabel('Death Rates per 100,000 population')
ax.set_ylabel('State')
ax.set_title('Death Rates due to Diseases vs Death Rates due to Accidents, Injury by firearms, Assault in the United States in 2006')
ax.legend()

ax.set_xticks([-150,-100,-50,0, 50, 100, 150, 200,250,300,350,400,450,500,550,600,650,700])
ax.set_xticklabels([abs(x) for x in ax.get_xticks()]) #Aboslute values are considered here to make sure the death rates numbers appear as positive

for state, diseases, accidents in zip(sorted_by_accidents['State'], sorted_by_accidents['Deaths due to Diseases'], sorted_by_accidents['Deaths due to Accidents,Injury by firearms,Assault']):
    ax.text(diseases + 2, state, f'{diseases:.1f}', ha='left', va='center', color='blue' )
    ax.text(-accidents - 5, state, f'{accidents:.1f}',ha='right', va='center', color='orange')

plt.show()
```

```python
df2=pd.read_csv('10s0107.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np


plt.figure(figsize=(12, 8))
for category in ["Male", "Female", "White", "Black"]:
    available_data = np.array(df2[category])
    mask = ~np.isnan(available_data)
    plt.plot(df2["Year"][mask], available_data[mask], label=category, marker='o')

plt.title("Infant Mortality Rates per 100,000 population between 1980-1999")
plt.xlabel("Year")
plt.ylabel("Infant Mortality Rate per 100,000 population")
plt.legend()
plt.xticks([1980,1985,1990,1995,1996,1997,1998,1999,2000])
plt.grid(True)
plt.show()
```
